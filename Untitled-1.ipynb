```python
import matplotlib.pyplot as plt
import numpy as np

# 初期値設定
z_0 = 0  # 高赤方偏移 z ≝ (z + 1)
Omega_Lambda0 = 0.68  # 暗黒エネルギー成分
Omega_r0 = 1e-4  # 放射成分
Omega_m0 = 1 - Omega_Lambda0 - Omega_r0  # 物質成分

h = 0.01  # ルンゲクッタ法のステップサイズ
num_steps = 4000  # 計算ステップ数

# 微分方程式の定義
def dOmega_rdN(Omega_r, Omega_Lambda):
    return Omega_r * (Omega_r - 3 * Omega_Lambda - 1)

def dOmega_LambdadN(Omega_r, Omega_Lambda):
    return Omega_Lambda * (Omega_r - 3 * Omega_Lambda + 3)

# 数値積分（ルンゲクッタ法）
def runge_kutta(Omega_r, Omega_Lambda, z, direction):
    z_list = [z + 1]
    Omega_r_list = [Omega_r]
    Omega_Lambda_list = [Omega_Lambda]
    Omega_m_list = [1 - Omega_Lambda - Omega_r]

    for _ in range(num_steps):
        k1_Omega_r = dOmega_rdN(Omega_r, Omega_Lambda)
        k1_Omega_Lambda = dOmega_LambdadN(Omega_r, Omega_Lambda)

        k2_Omega_r = dOmega_rdN(
            Omega_r + h * k1_Omega_r / 2,
            Omega_Lambda + h * k1_Omega_Lambda / 2,
        )
        k2_Omega_Lambda = dOmega_LambdadN(
            Omega_r + h * k1_Omega_r / 2,
            Omega_Lambda + h * k1_Omega_Lambda / 2,
        )

        k3_Omega_r = dOmega_rdN(
            Omega_r + h * k2_Omega_r / 2,
            Omega_Lambda + h * k2_Omega_Lambda / 2,
        )
        k3_Omega_Lambda = dOmega_LambdadN(
            Omega_r + h * k2_Omega_r / 2,
            Omega_Lambda + h * k2_Omega_Lambda / 2,
        )

        k4_Omega_r = dOmega_rdN(Omega_r + h * k3_Omega_r, Omega_Lambda + h * k3_Omega_Lambda)
        k4_Omega_Lambda = dOmega_LambdadN(
            Omega_r + h * k3_Omega_r,
            Omega_Lambda + h * k3_Omega_Lambda,
        )

        Omega_r += h * (k1_Omega_r + 2 * k2_Omega_r + 2 * k3_Omega_r + k4_Omega_r) / 6
        Omega_Lambda += h * (
            k1_Omega_Lambda + 2 * k2_Omega_Lambda + 2 * k3_Omega_Lambda + k4_Omega_Lambda
        ) / 6
        z = z * np.exp(direction * h)  # zの更新

        Omega_m = 1 - Omega_Lambda - Omega_r

        z_list.append(z + 1)
        Omega_r_list.append(Omega_r)
        Omega_Lambda_list.append(Omega_Lambda)
        Omega_m_list.append(Omega_m)

    return z_list, Omega_r_list, Omega_m_list, Omega_Lambda_list

# 増加方向 (z+1 = 10^0 → 10^6)
z_list_inc, Omega_r_list_inc, Omega_m_list_inc, Omega_Lambda_list_inc = runge_kutta(Omega_r0, Omega_Lambda0, z_0, 1)

# 減少方向 (z+1 = 10^0 → 10^-1)
z_list_dec, Omega_r_list_dec, Omega_m_list_dec, Omega_Lambda_list_dec = runge_kutta(Omega_r0, Omega_Lambda0, z_0, -1)

# FIG.1 のプロット
fig, ax = plt.subplots(figsize=(7, 7))

# 増加方向のプロット
ax.plot(z_list_inc, Omega_r_list_inc, label=r"$\Omega_r$ (increase)")
ax.plot(z_list_inc, Omega_m_list_inc, label=r"$\Omega_m$ (increase)")
ax.plot(z_list_inc, Omega_Lambda_list_inc, label=r"$\Omega_\Lambda$ (increase)")

# 減少方向のプロット
ax.plot(z_list_dec, Omega_r_list_dec, label=r"$\Omega_r$ (decrease)", linestyle="--")
ax.plot(z_list_dec, Omega_m_list_dec, label=r"$\Omega_m$ (decrease)", linestyle="--")
ax.plot(z_list_dec, Omega_Lambda_list_dec, label=r"$\Omega_\Lambda$ (decrease)", linestyle="--")

# 軸ラベルとスケール
ax.set_xscale("log")
ax.set_xlabel(r"$z+1$", fontsize=14)
ax.set_ylabel(r"$\Omega$", fontsize=14)
ax.set_xlim(10**-1, 10**6)
ax.set_ylim(0, 1.1)
ax.tick_params(which="both", direction="in", top=True, right=True)
ax.legend()

# 図の表示
plt.show()
```
